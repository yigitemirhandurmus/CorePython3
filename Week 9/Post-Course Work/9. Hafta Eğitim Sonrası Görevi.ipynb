# Problem 1

Kumanda isminde bir sınıf oluşturarak aşağıdaki işlemleri yapmasını sağlayan fonksiyonları sınıf içerisinde yazalım.
1. Tv Aç
2. Tv Kapat
3. Ses Ayarları
4. Kanal Ekle
5. Kanal Sayısını Öğrenme
6. Rastgele Kanala Geçme
7. Televizyon Bilgileri


Kumandayı kapatmak için q tuşuna basılmalı ve diğer durumlarda hangi işlemi yapmak istediği sorulmalıdır. Her işlemde gerekli fonksiyonunun sınıf içerisinden çağırılması gerekmektedir.

random ve time kütüphanesini araştırarak kullanabilirsiniz.

Kendi istediğiniz özellikleri eklemekte serbetsiniz.

```python
# 1.Problemin Çözümünü Buraya Yazınız

import random
import time

class Kumanda:
    def __init__(self):
        self.tv_durum = False
        self.ses = 50
        self.kanal_listesi = ["TRT 1", "ATV", "NTV", "SHOW", "KANAL D"]
        self.simdiki_kanal = self.kanal_listesi[0]

    def tv_ac(self):
        if self.tv_durum:
            print("TV zaten açık.")
        else:
            print("TV açılıyor...")
            self.tv_durum = True
            time.sleep(1)
            print("TV açıldı.")

    def tv_kapat(self):
        if not self.tv_durum:
            print("TV zaten kapalı.")
        else:
            print("TV kapanıyor...")
            self.tv_durum = False
            time.sleep(1)
            print("TV kapandı.")

    def ses_ayarlari(self):
        if not self.tv_durum:
            print("Önce TV'yi açın.")
            return
        while True:
            cevap = input("Sesi azalt: '<' Sesi artır: '>' Çıkış: 'q'\n")
            if cevap == "<":
                if self.ses > 0:
                    self.ses -= 10
                    print(f"Ses: {self.ses}")
            elif cevap == ">":
                if self.ses < 100:
                    self.ses += 10
                    print(f"Ses: {self.ses}")
            elif cevap == "q":
                print(f"Ses güncellendi: {self.ses}")
                break

    def kanal_ekle(self):
        if not self.tv_durum:
            print("Önce TV'yi açın.")
            return
        kanal = input("Eklemek istediğiniz kanalı girin: ")
        print("Kanal ekleniyor...")
        time.sleep(1)
        self.kanal_listesi.append(kanal)
        print(f"{kanal} kanalı eklendi.")

    def kanal_sayisi(self):
        if not self.tv_durum:
            print("Önce TV'yi açın.")
            return
        print(f"Kanal Sayısı: {len(self.kanal_listesi)}")

    def rastgele_kanal(self):
        if not self.tv_durum:
            print("Önce TV'yi açın.")
            return
        rastgele = random.choice(self.kanal_listesi)
        print(f"Şu anki kanal: {rastgele}")
        self.simdiki_kanal = rastgele

    def tv_bilgileri(self):
        print(f"""
        TV Durumu: {'Açık' if self.tv_durum else 'Kapalı'}
        Şu anki kanal: {self.simdiki_kanal if self.tv_durum else 'TV kapalı'}
        Ses: {self.ses if self.tv_durum else 'TV kapalı'}
        Kanal Listesi: {', '.join(self.kanal_listesi)}
        """)

print("""
Televizyon Uygulaması
1. TV Aç
2. TV Kapat
3. Ses Ayarları
4. Kanal Ekle
5. Kanal Sayısını Öğrenme
6. Rastgele Kanala Geçme
7. Televizyon Bilgileri
Çıkmak için 'q' tuşuna basın.
""")

kumanda = Kumanda()

while True:
    islem = input("İşlemi seçiniz: ")
    
    if islem == "q":
        print("Çıkış yapılıyor...")
        break
    elif islem == "1":
        kumanda.tv_ac()
    elif islem == "2":
        kumanda.tv_kapat()
    elif islem == "3":
        kumanda.ses_ayarlari()
    elif islem == "4":
        kumanda.kanal_ekle()
    elif islem == "5":
        kumanda.kanal_sayisi()
    elif islem == "6":
        kumanda.rastgele_kanal()
    elif islem == "7":
        kumanda.tv_bilgileri()
    else:
        print("Geçersiz işlem.")
```
