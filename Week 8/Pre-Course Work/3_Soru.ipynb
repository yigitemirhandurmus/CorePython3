Çeşitli metin dosyalarından karakterleri ayıklamak ve bir listeye koymak için bir Python programı yazın. ```test.txt``` ve ```demo.txt``` dosyalarından faydalanabilirsiniz.

````python
#Sorunun Çözmünü Buraya Gerçekleştirebilirsiniz

import os

if not os.path.exists("test.txt"):
    with open("test.txt", "w", encoding="utf-8") as file:
        pass

with open("test.txt", "r+", encoding="utf-8") as file:
    liste = []
    dosya = file.readlines()
    for x in dosya:
        for a in x:
            if a != "\n":
                liste.append(a)
    print(liste)

# OR

import glob

charList = []

fileList = glob.glob("test.txt")

for fileElements in fileList:
    with open(fileElements, "r") as f:
        charList.append(f.read())
print(charList)
````
