# Problem 1 
Kullanıcıdan aldığı ürün bilgisini (ad, fiyat) urunler.txt dosyasına kayıt eden fonksiyonu yazınız.

````python
# 1.Problemin Çözümünü Buraya Yazınız

import os

def writeItems():
    file.write("Ürün, Fiyat\n")

if not os.path.exists("urunler.csv"):
    with open("urunler.csv", "w", encoding="utf-8") as file:
        writeItems()

with open("urunler.csv", "r+", encoding="utf-8") as file:
    x = file.read()
    if not x.startswith("Ürün, Fiyat\n"):
        if x.startswith("Ürün, Fiyat"):
            file.write("\n")
        else:
            writeItems()

with open("urunler.csv", "r+", encoding="utf-8") as file:
    liste = []
    i = 1
    while True:
        ad = str(input(f"{i}. ürünün adını girin (Çıkmak için sadece Enter tuşuna basın): "))
        if ad == "":
            break
        fiyat = float(input(f"{i}. ürünün fiyatını girin: "))
        liste.append(f"{ad}, {fiyat}\n")
        i += 1
    
    mevcutIcerikler = file.read()
    yeniIcerikler = mevcutIcerikler + ''.join(liste)
    file.seek(0)
    file.write(yeniIcerikler)
````

# Problem 2
1.problemde oluşturulan text dosyası içerisinde kullanıcıdan alacağımız eski kelime isimli değişkene atadığımız 
ve değiştirilmesini istediğimiz kelimemizi yerini alıcak olan yeni kelime isimli değişkene atayın. Belirlenen kelimeleri text dosyasında değiştiren fonksiyonu yazınız.

````python
# 2.Problemin Çözümünü Buraya Yazınız

import csv

with open("urunler.csv", "r", encoding="utf-8") as file:
    reader = csv.reader(file, delimiter=',')
    lines = list(reader)

while True:
    old_name = input("Değiştirmek istediğiniz ürünün adını girin (Çıkmak için sadece Enter tuşuna basın): ")
    if old_name == "":
        break
    new_name = input("Yeni ürünün adını giriniz: ")
    
    for line in lines:
        if line[0] == old_name:
            line[0] = new_name
            print(f"Ürün adı '{new_name}' olarak değiştirildi.")
            break
    else:
        print(f"{old_name} bulunamadı.")

with open("urunler.csv", "w", encoding="utf-8", newline='') as file:
    writer = csv.writer(file, delimiter=',')
    writer.writerows(lines)
````

# Problem 3
Vehicle sınıfının tüm değişkenlerini ve yöntemlerini devralacak bir alt sınıf olarak Bus sınıfı oluşturunuz.



```
Output
> Name: School Volvo Speed: 180 Mileage: 12
```


````python
# 3.Problemin Çözümünü Buraya Yazınız


class Vehicle():
    
    def __init__(self, name, speed, mileage):
        self.name = name
        self.speed = speed
        self.mileage = mileage
        
    def show(self):
        print(f"Name: {self.name}, Speed: {self.speed}, Mileage: {self.mileage}")
 
class Bus(Vehicle):
    def show(self):
        pass

bus_instance = Bus("School Volvo", 180, 12)
 
print(f"Name: {bus_instance.name}, Speed: {bus_instance.speed}, Mileage: {bus_instance.mileage}")
````
